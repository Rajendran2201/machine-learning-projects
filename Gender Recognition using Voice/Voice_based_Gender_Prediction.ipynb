```python
# Splitting the dataset into features (X) and label (Y)
X = voice_data.drop('label', axis=1)
Y = voice_data['label']
```

```python
# Applying label mapping
def map_label(label):
    return 1 if label.lower() == 'male' else 0

Y = voice_data['label'].apply(map_label)
```

Scaling

```python
#Scaling

from sklearn.preprocessing import MinMaxScaler
# Scaling the features
scaler = MinMaxScaler((-1, 1))
X = scaler.fit_transform(X)
```

Splitting into train and test data

```python
# Splitting into train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Training the model

```python
# loading the support vector machine

from sklearn.svm import SVC # "Support Vector Classifier"
# Training the model
classification = SVC(kernel='linear')
classification.fit(X_train, Y_train)
```

```python
Y_prediction = classification.predict(X_test)
```

```python
print(Y_prediction)
```

```python
print("Accuracy score is",accuracy_score(Y_prediction,Y_test))
```
